```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
pd.merge(logins,registrations,on='name')
```

```python
pd.merge(registrations,logins,on='name',how='left')
```

```python
pd.merge(registrations,logins,on='name',how='right').fillna(0)
```

```python
pd.merge(registrations,logins,on='name',how='outer')
```

```python
registrations = registrations.set_index('name')
```

```python
registrations
```

```python
pd.merge(registrations,logins,left_index=True,right_on='name')
```

```python
registrations = registrations.reset_index()
```

```python
registrations
```

```python
registrations.columns = ["r_name","reg_id"]
```

```python
registrations
```

```python
pd.merge(registrations,logins,left_on='r_name',right_on='name')
```

```python
registrations.columns = ["name","id"]
logins.columns = ['id',"name"]
```

```python
pd.merge(registrations,logins,on='name',suffixes=("_reg","_log"))
```
