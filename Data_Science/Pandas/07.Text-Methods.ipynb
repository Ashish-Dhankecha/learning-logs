```python
messy_names = pd.Series(["andrew  ","bo;bo","   claire  "])
```

```python
messy_names
```

```python
messy_names.str.replace(';',"").str.strip().str.capitalize()
```

```python
def cleanup(name):
    name = name.replace(";",'')
    name = name.strip()
    name = name.capitalize()
    return name
```

```python
messy_names.apply(cleanup)
```
